```python
import requests
import json
from dadata import Dadata
```

```python
token = ""
```

```python
def create_company_report():
    global result
    # Запрашиваем ИНН у пользователя
    inn = input("Введите ИНН организации: ").strip()
    
    # Получаем данные
    dadata = Dadata(token)
    result = dadata.find_by_id("party", inn)
    
    if not result:
        print("Организация не найдена!")
        return
    
    # Берем первую компанию из результата
    company = result[0]['data']
    
    # Извлекаем нужные поля с проверкой на None
    name = company.get('name', {}).get('full_with_opf', 'Не указано') or 'Не указано'
    address = company.get('address', {}).get('value', 'Не указан') or 'Не указан'
    status = company.get('state', {}).get('status', 'Не указан') or 'Не указан'
    ogrn = company.get('ogrn', 'Не указан') or 'Не указан'
    okved = company.get('okved', 'Не указан') or 'Не указан'
    
    # Уставный капитал
    capital = company.get('capital', {})
    capital_value = capital.get('value', 'Не указан') if capital else 'Не указан'
    
    # Количество сотрудников
    employees = company.get('employee_count', 'Не указано') or 'Не указано'
    
    # Финансовые показатели
    finance = company.get('finance', {})
    finance_year = finance.get('year', 'Не указан') if finance else 'Не указано'
    finance_income = finance.get('income', 'Не указано') if finance else 'Не указано'
    
    # Учредители
    founders = company.get('founders', [])
    founders_list = ""
    if founders:
        for founder in founders:
            founder_name = founder.get('name', 'Неизвестно') or 'Неизвестно'
            founders_list += f"<li>{founder_name}</li>"
    else:
        founders_list = "<li>Не указаны</li>"
    
    # Создаем HTML
    html_content = f"""<html>
<head>
    <meta charset="UTF-8">
    <title>Отчет по ИНН {inn}</title>
</head>
<body>
    <h1>Отчет по организации</h1>
    <p><strong>ИНН:</strong> {inn}</p>
    <p><strong>Название:</strong> {name}</p>
    <p><strong>ОГРН:</strong> {ogrn}</p>
    <p><strong>Статус:</strong> {status}</p>
    <p><strong>Адрес:</strong> {address}</p>
    <p><strong>Уставный капитал:</strong> {capital_value}</p>
    <p><strong>Количество сотрудников:</strong> {employees}</p>
    <p><strong>Основной ОКВЭД:</strong> {okved}</p>
    <p><strong>Финансовые показатели за год:</strong> {finance_year}</p>
    <p><strong>Доходы:</strong> {finance_income}</p>
    <h2>Учредители:</h2>
    <ul>{founders_list}</ul>
</body>
</html>"""
    
    # Сохраняем в файл с указанием кодировки
    filename = f"report_{inn}.html"
    with open(filename, "w", encoding="utf-8") as f:
        f.write(html_content)
    
    print(f"Отчет сохранен в файл: {filename}")

# Запускаем функцию
create_company_report()
```
